```python
def create_tar_file(num_files_in_tar_file, path):
    os.chdir(path)
    tar = tarfile.open("size" + str(num_files_in_tar_file) + ".tar.gz", "w")
    for i in range(num_files_in_tar_file):
        name = "file" + str(i) + ".txt"
        with open(name, 'w') as f:
            f.write('r')
        tar.add(name)
        
        
```
