# Agglomerative Hierarchical 

```python
# compute distance matrix based on hamming
def trinary_distance_matrix(N_vectors):
    N = len(N_vectors)
    matrix = np.zeros((N,N))
    for i in range(N):
        for j in range(N):
            matrix[i,j] = sum(np.array(N_vectors[i]) != np.array(N_vectors[j])) / len(N_vectors[i])
    return matrix
```

```python
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering

def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)

```

```python
model = AgglomerativeClustering(distance_threshold=0, n_clusters = None, linkage = "average", affinity = "precomputed")
```

```python

# requires a df with columns searchlight, features and cluster assignment
def from_cluster_labels_to_fingerprint_avg(df, light_to_template_to_pid_to_cond_to_lists, condition_types):
    cluster_to_template_to_cond_to_list_of_lists = {}
    # for each cluster label
    for cluster_label in set(df["cluster_assignment"].tolist()):
        cluster_to_template_to_cond_to_list_of_lists[cluster_label] = {}
        # get all of the searchlights in this cluster label
        light_list = df[df["cluster_assignment"] == cluster_label]["searchlight"].tolist()
        # now we create list of lists where the list of lists contains lists in different searchlights
        # but with the sample cluster, template, pid (participant id), and condition
        for light_id in light_list:
            for template_id in light_to_template_to_pid_to_cond_to_lists[light_id]:
                if template_id not in cluster_to_template_to_cond_to_list_of_lists[cluster_label]:
                    cluster_to_template_to_cond_to_list_of_lists[cluster_label][template_id] = {}
                for pid in light_to_template_to_pid_to_cond_to_lists[light_id][template_id]:
                    for cond in light_to_template_to_pid_to_cond_to_lists[light_id][template_id][pid]:
                        if cond not in cluster_to_template_to_cond_to_list_of_lists[cluster_label][template_id]:
                            cluster_to_template_to_cond_to_list_of_lists[cluster_label][template_id][cond] = []
                        cluster_to_template_to_cond_to_list_of_lists[cluster_label][template_id][cond].append(light_to_template_to_pid_to_cond_to_lists[light_id][template_id][pid][cond])
    # previous we had a list of lists across searchlights
    # for each cluster, template and condition, get a vector of means across participants and searchlights
    cluster_to_template_to_cond_to_mean_list = {}
    for cluster in cluster_to_template_to_cond_to_list_of_lists:
        cluster_to_template_to_cond_to_mean_list[cluster] = {}
        for template in cluster_to_template_to_cond_to_list_of_lists[cluster]:
            cluster_to_template_to_cond_to_mean_list[cluster][template] = {}
            for cond in condition_types:
                this_cluster_template_condition_list_of_lists = cluster_to_template_to_cond_to_list_of_lists[cluster][template][cond]
                cluster_template_condition_matrix = np.vstack(this_cluster_template_condition_list_of_lists)
                cluster_to_template_to_cond_to_mean_list[cluster][template][cond] = np.nanmean(cluster_template_condition_matrix, axis = 0)
    return cluster_to_template_to_cond_to_mean_list
```
